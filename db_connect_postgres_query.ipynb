```python
import psycopg2
from google.colab import userdata

conn = psycopg2.connect(
    host='machine-learning-3253-postgresql.cluster-cg8zwc77lit0.ca-central-1.rds.amazonaws.com',
    port=5432,
    user='postgres',
    password=userdata.get('postgres.password'),
    dbname='lit',
)

print(conn)
```

```python
# docs: https://pandas.pydata.org/docs/reference/api/pandas.read_sql_query.html
import pandas.io.sql as sqlio
import pandas as pd
import timeit

def read_sql_query(q, conn):
    tic=timeit.default_timer()
    print(f"tic: {tic}")
    df = sqlio.read_sql_query(q, conn)
    toc=timeit.default_timer()
    print(f"toc: {toc}")
    print(f"benchmark: { toc - tic : .3f} seconds")
    return df
```

# Show full record count

```python
q = "select count(*) from lit_panel;"
df = read_sql_query(q, conn)
display(df)
```

# Count of records with `transactionDates` in 2022

```python
q = "select count(*) from lit_panel where transactionDate LIKE '2022-%'"
df = read_sql_query(q, conn)
display(df)
```

# Count of records with `transactionDates` in 2023

```python
q = "select count(*) from lit_panel where transactionDate LIKE '2023-%'"
df = read_sql_query(q, conn)
display(df)
```

# Count of records with `transactionDates` in 2024

```python
q = "select count(*) from lit_panel where transactionDate LIKE '2024-%'"
df = read_sql_query(q, conn)
display(df)
```

# Show top 10 `accessionNumber` with maximum multiplicity  

```python
q = "select accessionNumber, count(*) from lit_panel group by accessionNumber order by count(*) desc limit 10;"
df = read_sql_query(q, conn)
display(df)
```

```python
q = "select * from lit_panel where accessionNumber='0001019056-12-000683';"
df = read_sql_query(q, conn)
display(df)
```

# Show `LIKE '%TSLA%'` records

```python
q = "select * from lit_panel where issuerTradingSymbol LIKE '%TSLA%'"
df = read_sql_query(q, conn)
display(df)
```

# Use `= 'TSLA'` instead... (indexed now!)

```python
q = "select * from lit_panel where issuerTradingSymbol = 'TSLA'"
df = read_sql_query(q, conn)
display(df)
```
