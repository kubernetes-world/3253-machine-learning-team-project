```python
q = "select issuerTradingSymbol, count(*) from lit_panel group by issuerTradingSymbol order by count(*) desc limit 10;"
conn.sql(q).df()
```

```python
q = "select * from lit_panel where issuerTradingSymbol = 'GOOG'"
GOOG = conn.sql(q).df()
display(GOOG.columns)
display(GOOG)
```

```python
def drop_columns(dataframe):
  columns_to_drop = [col for col in dataframe.columns if col.endswith('Fn')]
  columns_to_drop.extend(['type', 'URL', 'dateOfFilingDateChange', 'issuerCIK',
                          'periodOfReport', 'dateOfOriginalSubmission',
                          'deemedExecutionDate', 'transactionFormType', 'valueOwnedFollowingTransaction',
                          'transactionTotalValue', 'exerciseDate', 'expirationDate',
                          'underlyingSecurityTitle', 'underlyingSecurityShares', 'underlyingSecurityValue'])
  return dataframe.drop(columns=columns_to_drop)
```

```python
GOOG_clean = drop_columns(GOOG)

display(GOOG_clean.columns)

# further drop more columns
GOOG_clean = GOOG_clean.drop(columns=[
    'documentType', 'notSubjectToSection16', 'issuerTradingSymbol', 'm_isDirector', 'm_isOfficer', 'm_isTenPercentOwner', 'm_isOther',
    'tableRow', 'transactionCode', 'transactionTimeliness'])

display(GOOG_clean)
```
