```python
import duckdb

conn = duckdb.connect('lit_panel.duckdb')
df = conn.sql('SELECT count(*) as tota_count FROM lit_panel').df()
df.head()
```

```python
q = "select count(*) as transactions_2022 from lit_panel where transactionDate LIKE '2022-%'"
df = conn.sql(q).df()
df
```

```python
q = "select count(*) as transactions_2023 from lit_panel where transactionDate LIKE '2023-%'"
df = conn.sql(q).df()
df
```

```python
q = "select count(*) as transactions_2024 from lit_panel where transactionDate LIKE '2024-%'"
df = conn.sql(q).df()
df
```

```python
q = "select accessionNumber, count(*) from lit_panel group by accessionNumber order by count(*) desc limit 10;"
df = conn.sql(q).df()
df
```

```python
q = "select * from lit_panel where accessionNumber='0001209191-04-053320';"
df = conn.sql(q).df()
df
```
