```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Replace 'path_to_your_file.csv' with the actual path to your CSV file
file_path = 'Apple_PriceHistory_simple.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y %I:%M %p')

# Extract month, day, and year into separate columns
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['year'] = df['Date'].dt.year

# Drop the original date column if you no longer need it
df.drop(columns=['Date'], inplace=True)
```

```python
# Display the first few rows of the DataFrame to verify
df.head(40)
```

```python
# Download the modified DataFrame to a new CSV file
df.to_csv('Apple_PriceHistory_modified.csv', index=False)
```
