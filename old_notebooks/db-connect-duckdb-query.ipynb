# 2023

```python
q = "select * from lit_panel where transactionDate LIKE '2023-%'"
df_2023 = conn.sql(q).df()
df_2023
```

```python
df_2023.columns
```

```python
df_2023.describe()
```

```python
# List of columns to drop (all columns ending with 'Fn')
columns_to_drop = [col for col in df_2023.columns if col.endswith('Fn')]
columns_to_drop.extend(['type', 'URL', 'dateOfFilingDateChange', 'issuerCIK',
                        'periodOfReport', 'dateOfOriginalSubmission', 'transactionDate',
                        'deemedExecutionDate', 'transactionFormType', 'valueOwnedFollowingTransaction',
                        'transactionTotalValue', 'exerciseDate', 'expirationDate',
                        'underlyingSecurityTitle', 'underlyingSecurityShares', 'underlyingSecurityValue'])

# Drop these columns from the DataFrame
df_2023_edited = df_2023.drop(columns=columns_to_drop)

# Now df_2023_edited will have all columns ending in 'Fn' removed
df_2023_edited.columns
```

# TOP `issuerTradingSymbol`

```python
q = "select issuerTradingSymbol, count(*) from lit_panel group by issuerTradingSymbol order by count(*) desc limit 10;"
df = conn.sql(q).df()
df
```

```python
q = "select issuerTradingSymbol, count(*) from lit_panel group by issuerTradingSymbol having count(*) = 100 order by count(*) desc;"
df_top_symbols_having_100 = conn.sql(q).df()
df_top_symbols_having_100
```

# Schema exploration

```python
# DO NOT DUMP THE ENTIRE DATAFRAME!!!

df = conn.sql('select * from lit_panel limit 1').df()
df.info()
```

```python
df.columns
```

```python
df.dtypes
```

```python
groups_by_col = {}
for col in [c for c in df.columns if not c.endswith('Fn') ]:
  print(f'GROUP BY column: {col}')
  query = f'SELECT {col}, count(*) FROM lit_panel GROUP BY {col} ORDER BY count(*) DESC;'
  df = conn.sql(query).df()
  groups_by_col[col] = df
  display(df)

display(groups_by_col)
```
